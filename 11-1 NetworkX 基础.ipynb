### 生成网络并可视化

```python
# 生成网络，命名为 G
G = nx.Graph()

# 加入节点 1 和 2 ，并产生连边
G.add_edge(1, 2)

# 节点标签
nx.draw(G, with_labels=True)
```

### 在网络中加入多条连边

```python
# 加入一群节点的连边
elist = [(2, 3), (1, 3), (1, 4)]
G.add_edges_from(elist)

nx.draw(G, with_labels=True)
```

### 生成有向网络并可视化

```python
# 有向网络
DG = nx.DiGraph()

DG.add_edge(1, 2)
DG.add_edges_from([(2, 3), (1, 3), (1, 4)])

nx.draw(DG, with_labels=True)
```

### 连边的属性

### 画图

```python
G = nx.cubical_graph()

# 网络的布局
pos1 = nx.circular_layout(G)
pos2 = nx.fruchterman_reingold_layout(G)

fig, axes = plt.subplots(nrows=1, ncols=2)

nx.draw(G, pos=pos1, ax=axes[0])
nx.draw(G, pos=pos2, node_color='r', edge_color='b', ax=axes[1])  # 节点、连边的颜色
```
